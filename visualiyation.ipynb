```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from math import ceil
# Initialize figure with subplots
def ploting_channels(df,ranges):
    fig= make_subplots(
        rows=4, cols=2, subplot_titles=("Channel 1", "Channel 2", "Channel 3", "Channel 4","Channel 5", "Channel 6", "Channel 7", "Channel 8")
    )
    row_ind=[1,1,2,2,3,3,4,4]
    for idx,rows in zip((range(0,8)),row_ind):
        # Add traces
        row_idx=rows
        if idx%2==0: col_idx=1
        else: col_idx=2
        fig.add_trace(go.Scatter(x=df["time_s"], y=df[idx] ),row=row_idx,col=col_idx)
        # Update xaxis properties
        fig.update_xaxes(title_text=f"time",  row=row_idx,col=col_idx)
        # Update yaxis properties
        fig.update_yaxes(title_text=f"{idx} channel", range=[-ranges, ranges], row=row_idx,col=col_idx)
        # Update title and height
        fig.update_layout(title_text="Example of 8 channel input signals", height=700)

    fig.show()
ploting_channels(li[0],1)
```
